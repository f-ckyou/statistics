## ðŸ§± Types of Statistical Graphs

Letâ€™s cover the **most common** and useful ones with examples:

| Type             | Purpose                               |
|------------------|----------------------------------------|
| Bar Chart        | Compare categories                     |
| Histogram        | Understand frequency distribution      |
| Pie Chart        | Show proportions                       |
| Box Plot         | Summarize data with 5-number summary   |
| Line Chart       | Show trend over time                   |
| Scatter Plot     | Explore relationships between variables|



```python
# BAR CHART
    # used for categorical data

categories = ['A', 'B', 'C']
values = [10, 24, 36]

plt.bar(categories, values)
plt.title("Bar Chart Example")
plt.xlabel("Category")
plt.ylabel("Values")
plt.show()
```

```python
# HISTOGRAM
    # used for continuous numerical data to show distribution

data = [13, 14, 15, 14, 13, 16, 18, 19, 17, 20, 21, 22]
plt.hist(data, bins=5, color='skyblue', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# PIE CHART
    # shows percentage/proportional data

labels = ['Apple', 'Banana', 'Cherry']
sizes = [30, 45, 25]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart Example")
plt.show()
```

```python
# BOX PLOT (BOX-and-WHISKER PLOT)
    # used to show 5-number summary: Min, Q1, Median, Q3, Max

data = [10, 12, 13, 15, 17, 18, 19, 21, 22, 23]
plt.boxplot(data)
plt.title("Box Plot Example")
plt.ylabel("Values")
plt.show()
```

```python
# LINE CHART
    # best for trends over time

years = [2018, 2019, 2020, 2021]
sales = [100, 150, 200, 250]

plt.plot(years, sales, marker='o')
plt.title("Sales Over Time")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
# SCATTER PLOT
    # used to show correlation or relationship between two variables

x = [1, 2, 3, 4, 5]
y = [5, 7, 9, 10, 15]

plt.scatter(x, y, color='purple')
plt.title("Scatter Plot Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

## ðŸŽ¨ Types of Data Visualization in Statistics (With Examples)

We can broadly classify them into the following categories:

---

### ðŸŸ¦ **1. Univariate Visualizations**
Visualizing a **single variable** â€” helps to understand **distribution**, **central tendency**, or **spread**.

| Type        | Purpose                       | Example                              |
|-------------|-------------------------------|--------------------------------------|
| Histogram   | Frequency of numerical data   | Height of students                   |
| Bar Chart   | Count of categories           | No. of students in each grade        |
| Pie Chart   | Proportions or percentages    | % of different blood groups          |
| Box Plot    | Spread and outliers           | Distribution of test scores          |
| Line Plot (Time series) | Change over time         | Temperature over days                |

---

### ðŸŸ¨ **2. Bivariate Visualizations**
Visualizing **two variables** â€” helps identify **relationships** or **comparisons**.

| Type         | Purpose                               | Example                          |
|--------------|----------------------------------------|----------------------------------|
| Scatter Plot | Relationship between variables         | Age vs Salary                    |
| Line Graph   | Trend of two variables over time       | Sales vs Months                  |
| Grouped Bar  | Compare values between subgroups       | Male vs Female across grades     |
| Hexbin Plot  | Density-based scatter alternative      | Large datasets with overlap      |

---

### ðŸŸ© **3. Multivariate Visualizations**
Visualize **three or more variables** â€” often uses **color**, **size**, or **position** to represent multiple dimensions.

| Type           | Purpose                                      | Example                             |
|----------------|-----------------------------------------------|-------------------------------------|
| Bubble Chart   | Like scatter plot + size                     | Revenue vs Profit vs Investment     |
| Heatmap        | Color-coded matrix                           | Correlation matrix between variables|
| Pair Plot      | Grid of scatter plots for many variables     | Full dataset exploration (Seaborn)  |
| 3D Plot        | Visualize 3D data                            | x, y, z values in physics simulation|

---

### ðŸŸ¥ **4. Statistical Summary Charts**
To summarize statistics (like distribution, quartiles, outliers).

| Type        | Purpose                           | Example                          |
|-------------|------------------------------------|----------------------------------|
| Box Plot    | 5-number summary                   | Exam score spread                |
| Violin Plot | Distribution + density             | Gender-wise income distribution  |
| ECDF Plot   | Empirical cumulative distribution  | Cumulative rainfall over days    |

---

### ðŸŸ« **5. Distribution & Density Visuals**
Show how data is spread or distributed.

| Type             | Purpose                           | Example                          |
|------------------|------------------------------------|----------------------------------|
| Histogram        | Frequency of values                | Marks distribution               |
| KDE Plot         | Smoothed density curve             | Income density in population     |
| Rug Plot         | Small lines for data positions     | Adds data points on axes         |

---

### ðŸŸª **6. Advanced/Interactive Visualizations**
Used for dashboards and complex datasets.

| Type           | Tool         | Description                           |
|----------------|--------------|----------------------------------------|
| Treemap        | Plotly/D3.js | Hierarchical proportion display        |
| Sunburst Chart | Plotly       | Multilevel pie-like chart              |
| Choropleth     | Plotly/GeoPandas | Geographic data (maps)            |
| Interactive Dashboards | Dash/Streamlit | For live visual reports     |

---

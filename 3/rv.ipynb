```python
import matplotlib.pyplot as plt
```

```python
# [ Bernoulli Distribution (Discrete) ]

from scipy.stats import bernoulli

# Define probability of success
p = 0.7
rv = bernoulli(p)

# PMF: Probability Mass Function
x = [0, 1]  # two possible outcomes
pmf = rv.pmf(x)

print("PMF values:", pmf)

# Plot PMF
plt.stem(x, pmf)
plt.title('Bernoulli Distribution (p = 0.7)')
plt.xlabel('x')
plt.ylabel('P(X=x)')
plt.xticks(x)
plt.grid(True)
plt.show()
```

```python
# [ Binomial Distribution ]

from scipy.stats import binom 

n = 10          # no of trials
p = 0.5         # success probability
rv = binom(n, p)

x = range(n + 1)
pmf = rv.pmf(x)

plt.bar(x, pmf)
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```

```python
# [ Poisson Distribution (Rare Events)]

from scipy.stats import poisson

#  = average number of events
lam = 3
rv = poisson(lam)

x = range(10)
pmf = rv.pmf(x)

plt.bar(x, pmf)
plt.title('Poisson Distribution ( = 3)')
plt.xlabel('Events')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```

```python
# [ Uniform Distribution (Continuous)]

from scipy.stats import uniform
import numpy as np

# Uniform between 0 and 1
a, b = 0, 1
rv = uniform(loc=a, scale=b-a)

x = np.linspace(a, b, 1000)
pdf = rv.pdf(x)

plt.plot(x, pdf)
plt.title('Uniform Distribution (0, 1)')
plt.xlabel('x')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
# [ Normal (Gaussian) Distribution ]

from scipy.stats import norm

# Mean = 0, Std dev = 1
mu, sigma = 0, 1
rv = norm(mu, sigma)

x = np.linspace(-4, 4, 1000)
pdf = rv.pdf(x)

plt.plot(x, pdf)
plt.title('Normal Distribution (=0, =1)')
plt.xlabel('x')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
# Exponential Distribution 

from scipy.stats import expon

#  = 1  scale = 1/ = 1
scale = 1
rv = expon(scale=scale)

x = np.linspace(0, 5, 1000)
pdf = rv.pdf(x)

plt.plot(x, pdf)
plt.title('Exponential Distribution ( = 1)')
plt.xlabel('x')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```
